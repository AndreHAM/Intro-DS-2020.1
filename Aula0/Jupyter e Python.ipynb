## Operações básicas

Python suporta as mesmas operações aritiméticas de C com algumas a mais.

````python
x = 2
y = 3

# Adição
print("Add:", y + x)
# Subtração
print("Sub:", y - x)
# Multiplicação
print("Mult:", y * x)
# Divisão real
print("Div:", y / x)
# Divisão inteira
print("Div int:", y // x)
# Exponenciação
print("Exp:", y ** x)

# Não exite x++
x += 1
print("x++ ", x)
````

````python
# AND
print(True and False)

# OR
print(True or False)

# Variaveis
print(var, y)
print(var > y)

# NOT
print(not True)
````

Perceba que True/False possuem a 1a letra em maiscula

```python
# odd numbers ^2, even numbers -> 0
l = []
for x in range(20):
    if x%2 != 0:
        l.append(x**2)
    else:
        l.append(0)
print(l)
```

````python
l = [x**2  if x%2 != 0 else 0 for x in range(20)] 
print(l)
````

## Dicionários


Quem não sabe o que é um dicionário?

## Dicionários

- São "tabelas" chave-valor
- Chave é única
- Mais rápido que percorrer uma lista
    - Acesso em O(1)
    - Tabela hash

````python
a = {'oi': 5, 'tchau': 10}

a
````

````python
a["oi"] = 0
a
````

É possível iterar sobre o dicionário da mesma forma como sobre as listas, só precisa escolher sobre o que 

````python
# Keys
print()

# Values 
print()

# Items
print()
````

Também temos comprehension em dicionários

````python
dt = {x:x**2 for x in range(10)}
print(dt)
````

## Função

Como sempre, é possível usar como se fosse em C mas o Python apresenta algumas funcionalidades a mais

```python
def funcao(parametro1, parametro2, ...):
    # codigo da funcao
```    

````python
# Função normal

def fun(a, b):
    return a * b

fun(4, 5)
````

````python
# Função que retorna 2 elementos
def fun2(a, b):
    pass
    # todo
fun2(1, 2)
````

````python
# Função com argumento padrão

def func3(a, b = 10):
    return a * b
````

````python
# Chamando normalmente
print(func3(5, 2))

# Sem passar o parametro
print(func3(5))
````
